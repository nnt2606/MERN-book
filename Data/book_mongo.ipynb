```python
book_dicts = book_df.to_dict('records')
book_dicts[0]
```

```python
id_dict = {k: v for k, v in id_df.values.tolist()}
len(id_dict)
```

```python
book_data = []

for book in book_dicts:
    book_id = book['BookID']
    book_series = None
    series = book['Series']
    
    if isinstance(series, str) and series != '':
        book_series = series
    
    book_rec = {
        'title': book['Title'],
        'authors': book['Author'].split(', '),
        'series': book_series,
        'genre': genre_df[genre_df['bookId'] == book_id].Genre.tolist()
    }
    
    if isinstance(series, str) and series != '':
        book_rec['series'] = series
    book_data.append(book_rec)
    
len(book_data)
```

```python
book_res = shop_db.books.insert_many(book_data)
res_id = book_res.inserted_ids
len(res_id)
```
